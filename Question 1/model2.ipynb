```python
from pyspark.sql import SparkSession

spark = SparkSession.builder\
        .master("local")\
        .appName("Colab")\
        .config('spark.ui.port', '4050')\
        .getOrCreate()
```

```python
# read data
df=spark.read \
  .option("header","True")\
  .option("inferSchema","True")\
  .option("sep",",")\
  .csv("DataFile.csv")


print("There are ",df.count(),"rows",len(df.columns),"columns","in data")
```

```python
# check data type
df.printSchema()
```

```python
# check data type
df.select("type").distinct().show(100)
```

```python
help(df.filter)
```

```python
# remove all null values
df = df.filter("type != \"null\"")
df = df.filter("repository_forks != \"null\"")
df = df.filter("repository_size != \"null\"")
df = df.filter("repository_open_issues != \"null\"")
df = df.filter("repository_watchers != \"null\"")
df1 = df.select("repository_forks","repository_size","repository_open_issues","repository_watchers","type")
```

```python
df.show(100)
df1.printSchema()
print(df1.filter("type == \"null\""))
```

```python
# convert null value to 0
from pyspark.sql.functions import when, lit, col, expr
df1 = df1.withColumn("repository_size",
                                  when(df1.repository_size.isNull() ,0)
                                 .otherwise(df1.repository_size))
df1 = df1.withColumn("repository_forks",
                                  when(df1.repository_forks.isNull() ,0)
                                 .otherwise(df1.repository_forks))
df1 = df1.withColumn("repository_open_issues",
                                  when(df1.repository_open_issues.isNull() ,0)
                                 .otherwise(df1.repository_open_issues))
df1 = df1.withColumn("repository_watchers",
                                  when(df1.repository_watchers.isNull() ,0)
                                 .otherwise(df1.repository_watchers))
```

```python
from pyspark.sql.functions import when, lit, col, expr
df1.show(30)
```

```python
# convert datatype to float
from pyspark.sql.functions import when, lit, col, expr
from pyspark.sql.types import IntegerType, FloatType
df1 = df1.withColumn("repository_forks",col("repository_forks").cast(FloatType()))
df1 = df1.withColumn("repository_size",col("repository_size").cast(FloatType()))
df1 = df1.withColumn("repository_open_issues",col("repository_open_issues").cast(FloatType()))
df1 = df1.withColumn("repository_watchers",col("repository_watchers").cast(FloatType()))
df1.printSchema()
```

```python
df1.select('type').distinct().show()
```

```python
from pyspark.ml.linalg import Vectors 
from pyspark.ml import Pipeline
from pyspark.ml.feature import IndexToString,StringIndexer, VectorIndexer
from pyspark.ml.tuning import CrossValidator, ParamGridBuilder
from pyspark.ml.evaluation import MulticlassClassificationEvaluator

#Transfer the data to make sure the machine learning algorithm can recongize them
def transData(data):
  return data.rdd.map(lambda r: [Vectors.dense(r[:-1]),r[-1]]).toDF(['features','label'])
```

```python
transformed = transData(df1)
transformed.show()
```

```python
transformed.select('label').distinct().show()
```

```python
#Index labels
labelIndexer = StringIndexer(inputCol='label',
                             outputCol='indexedLabel').fit(transformed)
labelIndexer.transform(transformed).show(30, True)
```

```python
# convert label into number
stringIndexer = StringIndexer()\
                .setInputCol('type')\
                .setOutputCol('label')
mymodel = stringIndexer.fit(df1)
df3=mymodel.transform(df1)
df3.show()
```

```python
# index features
featureIndexer =VectorIndexer(inputCol="features", \
                              outputCol="indexedFeatures", \
                              maxCategories=4).fit(transformed)
featureIndexer.transform(transformed)
```

```python
# Split the data into training and test
trainingData, testData = transformed.randomSplit([0.6, 0.4],seed=2022)

trainingData.show(5)
testData.show(5,False)
```

```python
from pyspark.ml.classification import LogisticRegression
logr = LogisticRegression(featuresCol='indexedFeatures', labelCol='indexedLabel', maxIter=10)
```

```python
# Convert indexed labels back to original labels.
labelConverter = IndexToString(inputCol="prediction", outputCol="predictedLabel",
                               labels=labelIndexer.labels)
```

```python
# Ccreate Pipeline
pipeline = Pipeline(stages=[featureIndexer, labelIndexer, logr, labelConverter])
```

```python
# Train model
model = pipeline.fit(trainingData)
```

```python
# Make predictions.
predictions = model.transform(testData)
# Select example rows to display.
predictions.show(500)
```

```python
from pyspark.ml.evaluation import MulticlassClassificationEvaluator

# do evaluation and compute test error
evaluator = MulticlassClassificationEvaluator(
    labelCol="indexedLabel", predictionCol="prediction", metricName="accuracy")
accuracy = evaluator.evaluate(predictions)
print("Test Error = %g" % (1.0 - accuracy))
```
