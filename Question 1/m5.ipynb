```python
from pyspark.sql.functions import when, lit, col, expr
from pyspark.sql.types import IntegerType, FloatType, StringType

print(df.count())
# Data Cleaning
df = df.withColumn("repository_has_issues",col("repository_has_issues").cast(StringType()))
df1 = df.filter(("repository_fork == \"TRUE\" OR repository_fork == \"FALSE\" ")).filter("type != \"null\"")
df1 = df1.select("repository_forks","repository_size","repository_open_issues","repository_watchers","repository_has_issues","repository_fork","type")
print(df1.count())
```

```python
# convert null value into 0
from pyspark.sql.functions import when, lit, col, expr
df1 = df1.withColumn("repository_size",
                                  when(df1.repository_size.isNull() ,0)
                                 .otherwise(df1.repository_size))
df1 = df1.withColumn("repository_forks",
                                  when(df1.repository_forks.isNull() ,0)
                                 .otherwise(df1.repository_forks))
df1 = df1.withColumn("repository_open_issues",
                                  when(df1.repository_open_issues.isNull() ,0)
                                 .otherwise(df1.repository_open_issues))
df1 = df1.withColumn("repository_watchers",
                                  when(df1.repository_watchers.isNull() ,0)
                                 .otherwise(df1.repository_watchers))
```

```python
# covert data type into float
from pyspark.sql.functions import when, lit, col, expr
from pyspark.sql.types import IntegerType, FloatType
from pyspark.ml.feature import OneHotEncoder,StringIndexer

df1 = df1.withColumn("repository_forks",col("repository_forks").cast(FloatType()))
df1 = df1.withColumn("repository_size",col("repository_size").cast(FloatType()))
df1 = df1.withColumn("repository_open_issues",col("repository_open_issues").cast(FloatType()))
df1 = df1.withColumn("repository_watchers",col("repository_watchers").cast(FloatType()))

# Transform repository_fork to the right format/encoding
indexr1 = StringIndexer() \
      .setInputCol("repository_fork") \
      .setOutputCol("repository_fork_id")

TempModel=indexr1.fit(df1)
df1_temp = TempModel.transform(df1)

encoder1 = OneHotEncoder() \
      .setInputCol("repository_fork_id") \
      .setOutputCol("repository_fork_encode")

TempModel2=encoder1.fit(df1_temp)
df1 = TempModel2.transform(df1_temp)

# Transform repository_has_issues to the right format/encoding
indexr1 = StringIndexer() \
      .setInputCol("repository_has_issues") \
      .setOutputCol("repository_has_issues_id")

TempModel=indexr1.fit(df1)
df1_temp = TempModel.transform(df1)

encoder1 = OneHotEncoder() \
      .setInputCol("repository_has_issues_id") \
      .setOutputCol("repository_has_issues_encode")

TempModel2=encoder1.fit(df1_temp)
df1 = TempModel2.transform(df1_temp)

df1.printSchema()

print(df1.show(20))
df1 = df1.select("repository_forks","repository_size","repository_open_issues","repository_watchers","repository_has_issues_encode","repository_fork_encode","type")
```

```python
from pyspark.ml.feature import VectorAssembler
df1.select('type').distinct().show()
#df1.printSchema()
#print(df1.show(194))
df1.select("repository_fork_encode").distinct().show(100)
df1.select("repository_has_issues_encode").distinct().show(100)

#Transform the data into the vectors
assembler = VectorAssembler() \
        .setInputCols(["repository_forks", "repository_size", "repository_open_issues", "repository_watchers", "repository_has_issues_encode", "repository_fork_encode"]) \
        .setOutputCol("vectorized_feature")

assembler_df = assembler.transform(df1)
print(assembler_df.show(30))
assembler_df.printSchema()
assembler_df.toPandas().head()

```

```python
from pyspark.ml.linalg import Vectors 
from pyspark.ml import Pipeline
from pyspark.ml.feature import IndexToString,StringIndexer, VectorIndexer
from pyspark.ml.tuning import CrossValidator, ParamGridBuilder
from pyspark.ml.evaluation import MulticlassClassificationEvaluator
```

```python
# index labels
labelIndexer = StringIndexer(inputCol='type',
                             outputCol='indexedLabel').fit(assembler_df)
labelIndexer.transform(assembler_df).show(30, True)
```

```python
#convert type into numbers
stringIndexer = StringIndexer()\
                .setInputCol('type')\
                .setOutputCol('label')
mymodel = stringIndexer.fit(assembler_df)
df3=mymodel.transform(assembler_df)
df3.show()
```

```python
# index features
featureIndexer =VectorIndexer(inputCol="vectorized_feature", \
                              outputCol="indexedFeatures", \
                              maxCategories=4).fit(df3)
featureIndexer.transform(df3)
```

```python
# Split the data into training and test
trainingData, testData = df3.randomSplit([0.6, 0.4],seed=2022)

trainingData.show(5)
testData.show(5,False)
```

```python
from pyspark.ml.classification import LogisticRegression

logistic = LogisticRegression(featuresCol='vectorized_feature', labelCol='label', maxIter=10)
```

```python
# try Decision Tree Classification Model
from pyspark.ml.classification import DecisionTreeClassifier

# Train a DecisionTree model
decisionTree = DecisionTreeClassifier(labelCol='label', featuresCol='vectorized_feature')
```

```python
# decision tree
treelabelConverter = IndexToString(inputCol="prediction", outputCol="predictedLabel",
                               labels=labelIndexer.labels)
```

```python
# create Pipeline
pipeline = Pipeline(stages=[featureIndexer, labelIndexer, decisionTree, treelabelConverter])
```

```python
# Train model
model = pipeline.fit(trainingData)


```

```python
# Make predictions.
predictions = model.transform(testData)
# Select example rows to display.
predictions.show()
```

```python
from pyspark.ml.evaluation import MulticlassClassificationEvaluator

# do eveluation and compute test error
evaluator = MulticlassClassificationEvaluator(
    labelCol="indexedLabel", predictionCol="prediction", metricName="accuracy")
accuracy = evaluator.evaluate(predictions)
print("Test Error = %g" % (1.0 - accuracy))
```
